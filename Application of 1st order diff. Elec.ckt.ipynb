## Charging of capacitor

V = VC + VL



or, V = Q/C + $R\frac{dQ}{dt}$

or, $\frac{dQ}{dt} = \frac{V}{R} -\frac{Q}{CR}$

```python
import numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
```

```python
def model(Q,t,C,R,V):
    dQdt = V/R- Q/(C*R)
    return dQdt
```

```python

V=50 #V
C=10**-6 #F
R1=100 #ohm
R2 = 200
R3=500
R4 = 700
R5 = 900
```

```python
# time points
t = np.arange(0,0.002,0.0001) 
```

```python
# Charging initial condition
Q0 = 0
Q1 = odeint(model,Q0,t,args=(C,R1,V))
Q2 = odeint(model,Q0,t,args=(C,R2,V))
Q3 = odeint(model,Q0,t,args=(C,R3,V))
Q4 = odeint(model,Q0,t,args=(C,R4,V))
Q5 = odeint(model,Q0,t,args=(C,R5,V))
```

```python
# plot results
plt.plot(t,Q1,'r-',linewidth=2,label='R1=100$\Omega$')
plt.plot(t,Q2,'b--',linewidth=2,label='R2=200$\Omega$')
plt.plot(t,Q3,'g.',linewidth=2,label='R3=500$\Omega$')
plt.plot(t,Q4,'c:',linewidth=2,label='R2=700$\Omega$')
plt.plot(t,Q5,'m^',linewidth=2,label='R3=900$\Omega$')
plt.xlabel('Time(sec)')  # $\Omega$ = unit of resistance
plt.xticks(rotation= 50)
plt.ylabel('Charge(C)')

plt.title("Charging of capacitor")
plt.legend()
plt.show()
```

## Discharging of Capacitor

```python
# Discharging initial condition
Q0 = V*C     # Q0 = Q max = V*C
Q11 = odeint(model,Q0,t,args=(C,R1,0))  # V = 0 in case of discharging 
Q22 = odeint(model,Q0,t,args=(C,R2,0))
Q33 = odeint(model,Q0,t,args=(C,R3,0))
Q44 = odeint(model,Q0,t,args=(C,R4,0))
Q55 = odeint(model,Q0,t,args=(C,R5,0))
```

```python
# plot results
plt.plot(t,Q11,'r-',linewidth=2,label='R1=100$\Omega$')
plt.plot(t,Q22,'b--',linewidth=2,label='R2=200$\Omega$')
plt.plot(t,Q33,'g.',linewidth=2,label='R3=500$\Omega$')
plt.plot(t,Q44,'c:',linewidth=2,label='R4=700$\Omega$')
plt.plot(t,Q55,'m^',linewidth=2,label='R5=900$\Omega$')
plt.xlabel('Time(sec)')  # $\Omega$ = unit of resistance
plt.xticks(rotation= 50)
plt.ylabel('Charge(C)')

plt.title("Discharging of capacitor")
plt.legend()
plt.show()
```
